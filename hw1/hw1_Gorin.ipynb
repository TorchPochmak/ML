7. Рассмотрим случайную матрицу  $15 × 2 $, представляющую декартовы координаты, преобразуйте их в полярные координаты.

```python
arr = np.random.random((15, 2))
x, y = arr.T
print("Input: ", arr)
r = np.hypot(x, y) 
theta = np.arctan2(y, x)
polar_coords = np.c_[r, theta] 
print("Output: ", polar_coords)
```

9. Рассмотрим матрицу $10 × 3 $, извлеките строки с неравными значениями (например, $[2,2,3]$).

```python
mn = 0
mx = 10
matr = np.random.randint(mn, mx, (10, 3))
matr[4] = np.ones(3, dtype=int)
unique = matr[~np.all(matr == matr[:, [0]], axis=1)]

print("Input: ", matr)
print("Output: ", unique)
```

# Найдите ответы на вопросы с помощью предложенного датасета:

Файл с датасетом - `olympics_dataset.csv`

```python
import pandas as pd
```

```python
# Load dataset here
df = pd.read_csv('olympics_dataset.csv')
df.tail(10)
```

1. Сколько спортсменов участвовали в Летних Олимпийских играх за всё время?

```python
summer_df = df[df['Season'] == 'Summer']['Name']
print("Число спортсменов  на летних Олимпийских играх: ", len(summer_df.unique()))
```

2. Найдите участника с наибольшим количеством медалей

```python
medals_count = df[df['Medal'] != 'No medal'].groupby('Name')['Medal'].count()
top_athlete = medals_count.idxmax()
top_athlete_medals = medals_count.max()
print("Участник с наибольшим количеством медалей:", top_athlete)
print("Число: ", top_athlete_medals)
```

3. Какой спорт является самым популярным, то есть в нём участвовало больше всего спортсменов за всё время?

```python
# Наиболее популярный спорт
popular = df.groupby('Sport')['Name'].nunique().idxmax()
print("Самый популярный спорт", popular)
```

4. Кто был победителем медального зачёта в 2004 году?

```python
medals = df[(df['Year'] == 2004) & (df['Medal'] != 'No medal')]
medals_country = medals.groupby('Team')['Medal'].count()
print("Победитель медального зачёта в 2004 году:", medals_country.idxmax())
```

5. В каком виде спорта США завоевало наибольшее число медалей?

```python
medals = df[(df['Team'] == 'United States') & (df['Medal'] != 'No medal')].groupby('Sport')['Medal'].count()
print("Спорт, в котором США завоевали наибольшее число медалей: ", medals.idxmax())
```

# С помощью графика покажите:

6. Сколько женщин и мужчин участвовало в Летних Олимпийских играх каждый год?

```python
import matplotlib.pyplot as plt
```

```python
summer = df[df['Season'] == 'Summer']
gender_counts = summer.groupby(['Year', 'Sex'])['Name'].nunique().unstack()
print(gender_counts.head(5))
plt.plot(gender_counts.index, gender_counts['F'], label='Женщины')
plt.plot(gender_counts.index, gender_counts['M'], label='Мужчины')

plt.xlabel('Год')
plt.ylabel('Количество спортсменов')
plt.legend(title='Пол')
plt.show()
```

7. Какие 10 стран на текущий момент имеют наибольшее количество золотых медалей?


```python
gold_medals = df[(df['Medal'] == 'Gold')].groupby('Team')['Medal'].count().sort_values(ascending=False).head(50)
plt.figure(figsize=(12, 6))
gold_medals.plot(kind='bar', color='blue', width=0.7)

plt.xlabel('Страна')
plt.xticks(rotation=90)
plt.margins(x=0.5) 
plt.ylabel('Количество золотых медалей')
plt.show()
print(gold_medals.head(10))
```

8. Сколько золотых, серебряных и бронзовых медалей зарабатывал СССР отдельно по каждым играм?

```python
ussr_medals = df[(df['Team'] == 'Soviet Union') & (df['Medal'] != 'No medal')]
medals_table = ussr_medals.pivot_table(index='Year', columns='Medal', aggfunc='size')
medals_table = medals_table[['Bronze','Silver', 'Gold']]
print(medals_table)
medals_table.plot(kind='bar', stacked=True, figsize=(12, 6), color=['brown', 'grey', 'yellow'])
```

9. Сколько медалей завоёвывали спортсмены и спортсменки (отдельно) команды Китая на каждой олимпиаде?


```python
china_medals = df[(df['Team'] == 'China') & (df['Medal'] != 'No medal')]
china_medals_counts = china_medals.groupby(['Year', 'Sex']).size().unstack()
china_medals_counts = china_medals_counts[['F', 'M']]
print(china_medals_counts)
china_medals_counts.plot(kind='bar', stacked=True, figsize=(12, 6), color=['pink', 'blue'])
```

10. Какие 10 стран имеют наибольшее количество участников, не получивших никаких медалей ни на одной из Летних олимпиад?

```python
all_participations = df[df['Season'] == 'Summer'][['Name', 'Team']]
no_medals = df[(df['Medal'] == 'No medal') & (df['Season'] == 'Summer')][['Name', 'Team']]
no_medals['cnt'] = no_medals.groupby('Name')['Name'].transform('count')
no_medals = no_medals.drop_duplicates().groupby('Team')['Name'].nunique().sort_values(ascending=False)

top_10_countries = no_medals.head(10)

print("Топ-10 стран с наибольшим количеством участников, не завоевавших медалей:\n", top_10_countries)
top_50_countries = no_medals.head(50)
plt.figure(figsize=(16, 8))
top_50_countries.plot(kind='bar')
plt.title("Топ-50 стран с наибольшим количеством участников, не завоевавших медалей")
plt.xlabel("Количество участников")
plt.ylabel("Страна")
plt.tight_layout()
plt.show()
```
